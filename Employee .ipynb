```python
crosstab = pd.crosstab(df['OverTime'], df['Attrition'])
crosstab 
```

```python
crosstab_percentage = crosstab.div(crosstab.sum(axis=1), axis=0)
crosstab_percentage
```

```python
# Plot
crosstab_percentage.plot(kind='bar', stacked=True, color=['blue', 'red'])
plt.xlabel('OverTime')
plt.ylabel('Percentage')
plt.title('Percentage Stacked Bar Chart')
plt.legend(['No Attrition', 'Attrition'])
plt.show()
```
